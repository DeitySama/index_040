```python
import pandas as pd;
import matplotlib.pyplot as plt;
import numpy as py;

df=pd.read_csv('./red_wine.csv');

#Print first ten rows
print("Printing the first ten rows", end='\n')
print(df.head(10))

#Print last twenty rows
print("Printing the last twenty rows", end='\n')
print(df.tail(20))

#Print Info
print("Using the info method", end='\n')
print(df.info())

#Print Describe
print("Using the describe method", end='\n')
print(df.describe())

#PANDAS Dataframe
ds = pd.DataFrame(df,columns=["fixed acidity","alcohol"]);

arr = ds.to_numpy();

print(arr[0])

plt.plot(arr[:, 0], arr[:, 1], marker='o', linestyle='')
plt.title('Matplotlib Plot of Pandas DataFrame')
plt.xlabel('Fixed Acidity')
plt.ylabel('Alcohol')
plt.show()
```
